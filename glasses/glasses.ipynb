## Task condition

Implement a solution to glasses dropping problem.

Given:

- number of floors
- number of glasses

Determine the minimum number of attempts required to find the critical floor (the highest floor from which a glass does not break), considering both outcomes of each drop:

- glass breaks
- glass does not break

Compute results for:

- 10 floors, 2 glasses
- 100 floors, 2 glasses
- 100 floors, 3 glasses
- 1000 floors, 3 glasses
- 1000 floors, 5 glasses

## Answer

```python
def max_floors(k, t):
    return 0 if k == 0 or t == 0 else (t if k == 1 else max_floors(k - 1, t - 1) + max_floors(k, t - 1) + 1)

def min_k(f, t):
    k = 0
    while max_floors(t, k) < f:
        k += 1
    return k

print("For 10 floors and 2 glasses:", min_k(10, 2))
print("For 100 floors and 2 glasses:", min_k(100, 2))
print("For 100 floors and 3 glasses:", min_k(100, 3))    
print("For 1000 floors and 3 glasses:",min_k(1000, 3))
print("For 1000 floors and 5 glasses:", min_k(1000, 5))
```
